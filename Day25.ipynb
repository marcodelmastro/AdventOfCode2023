```python
import networkx as nx

def readInput25(infile):
    with open(infile) as f:
        G = nx.Graph()
        graph = {}
        for l in f.read().strip().splitlines():
            i = l.split(": ")
            graph[i[0]] = i[1].split()
            for j in i[1].split():
                G.add_edge(i[0],j)
        return graph, G
```

```python
import matplotlib.pyplot as plt

_, Gtest = readInput25("examples/example25.txt")

nx.draw(Gtest, with_labels=True)
```

```python
Gtestcut = Gtest

Gtestcut.remove_edge("hfx","pzl")
Gtestcut.remove_edge("nvd","jqt")
Gtestcut.remove_edge("cmg","bvb")

nx.draw(Gtestcut, with_labels=True)
```

```python
prod = 1
for c in nx.connected_components(Gtestcut):
    print(len(c))
    prod *= len(c)
print(prod
```
